```python
# example from https://github.com/DLR-RM/stable-baselines3

import gym
from gym import wrappers
from stable_baselines3 import PPO

def vid(episode):
    return episode%20 == 0

env = gym.make("LunarLanderContinuous-v2")
env = wrappers.Monitor(env, "./gym-resultsPPO", force=True,video_callable = vid)

model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=500000, reset_num_timesteps=100)
```

```python
# run the last episode until finish
obs, reward, done, info =  env.step([0,0])
while not done:
    obs, reward, done, info =  env.step([0,0])



# get them sweet renders
obs = env.reset()
for i in range(1000):
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    env.render()
    if done:
        obs = env.reset()
```
